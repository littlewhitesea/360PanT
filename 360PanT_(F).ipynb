```python
#@title ## 3. Generate translated results
%cd /content/360PanT/freecontrol

#@markdown method choice: 1. FreeControl; 2. 360PanT_F
method = "360PanT_F" #@param {'type':'string'}
path_of_360panorama = "/content/360PanT/freecontrol/dataset/stuttgart_suburbs.jpg" #@param {'type':'string'}
#@markdown condition type choice: 1. canny; 2. seg; 3. depth; 4. None

#@markdown The other control condition (e.g., Canny Map, Segmentation Map or Depth Map) is directly extracted from the given 360-degree panorama, when condition type is None, the input 360-degree panoramic image is as the control condition.
condition_type = "seg" #@param {'type':'string'}
target_prompt = "oil painting of a photo of houses in the suburbs" #@param {'type':'string'}
original_prompt = "a photo of houses in the suburbs" #@param {'type':'string'}
key_word = "houses" #@param {'type':'string'}
#@markdown semantic base choice: 1. indoors; 2. outdoors
semantic_base = "outdoors" #@param {'type':'string'}
full_semantic_base = semantic_base + "_step_200_sample_20_id_0"
paired_objs = f"({key_word}; {key_word})"



command_line = f"python inference.py --approach {method} --condition_image {path_of_360panorama} --prompt '{target_prompt}' --pca_basis {full_semantic_base} --inversion_prompt '{original_prompt}' --condition '{condition_type}' --paired_objs \"{paired_objs}\""
# print(command_line)
!{command_line}
```
